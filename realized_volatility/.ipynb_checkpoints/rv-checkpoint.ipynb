```python
#Pick day
subset = dat.loc[dat.DATE >= '2016-06-20',:]
subset = subset.loc[subset.DATE < '2016-06-30',:]

#subset = dat
```

```python
subset.head()
```

```python
subset.TIME_M = subset.TIME_M.astype('datetime64[ns]')
```

```python
#Pick between 9:30 and 4
subset = subset.loc[subset.TIME_M >= '9:35:00.0',:]
subset = subset.loc[subset.TIME_M <= '16:00:00.0',:]
```

```python
subset.tail()
```

```python
subset['hour'] = subset.TIME_M.dt.hour
subset['minute'] = subset.TIME_M.dt.minute
```

```python
# Now grab only minutes div by 5
```

```python
subset.head()
```

```python
subset['mod_5'] = subset.TIME_M.dt.minute % 5
```

```python
subset = subset.loc[subset['mod_5'] == 0]
```

```python
# Remove prices equal to zero
subset = subset.loc[subset['ASK'] != 0]
subset = subset.loc[subset['BID'] != 0]
```

```python
subset.head()
```

```python
#subset.groupby(['hour','minute']).nth(0)
```

```python
subset.columns
```

```python
gr_by = subset.groupby(['DATE','hour','minute']).min('TIME_M')
```

```python
#gr_by = subset.groupby(['DATE','hour','minute'])['TIME_M'].min()
```

```python
gr_by.tail(n = 50)
```

```python
gr_by['log_PRICE'] = np.log(gr_by['PRICE'])
```

```python
gr_by['difflog_PRICE'] = gr_by['log_PRICE'].diff()
```

```python
gr_by['sq_difflog_PRICE'] = gr_by['difflog_PRICE']**2
```

```python
gr_by.head()
```

```python
gr_by.dropna(subset=['sq_difflog_PRICE'], inplace=True)
```

```python
gr_by.tail()
```

```python
gr_by.groupby(['DATE'])['sq_difflog_PRICE'].sum()
```

```python
plt.plot(gr_by['sq_difflog_PRICE'].reset_index()['sq_difflog_PRICE'])
plt.show()
```
